```python
from googleapiclient.discovery import build
import pandas as pd
```

```python

file_path = '/content/drive/Shareddrives/CSCE 470 Project/top_1000_youtubers.csv'

df = pd.read_csv(file_path)

urls = df['Youtube Url']
urls_df = pd.DataFrame(urls)
urls_df

urls_df['Youtube Url'] = urls_df['Youtube Url'].str.extract(r'channel/(.+)')

chan_ids = urls_df
chan_ids.to_csv('1000_chanIDs.csv', index=False)
chan_names = df['username']
chan_names_ids = pd.concat([chan_names, chan_ids], axis=1)
chan_names_ids.to_csv('1000_chan_names_IDs.csv', index=False)
```

```python

api_key = 'AIzaSyDGs6ef9SbzL5Bg2ioftdBL0BrPTpRgrwk'


youtube = build('youtube',
                'v3',
                developerKey=api_key,
)
# add video id to master dataframe
df['Video ID'] = ''

for index, row in chan_ids.iterrows():
  if index < 347:
    continue
  #make a req
  request = youtube.search().list(
          part="snippet",
          channelId=row['Youtube Url'],
          maxResults=1,
          order="date",
  )

  response = request.execute()

  result_type = (response['items'][0]['id']['kind'])
  if result_type == "youtube#video":
      video_id = response['items'][0]['id']['videoId']
      df.at[index,'Video ID'] = video_id


```
